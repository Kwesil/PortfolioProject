```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Setting our data
df_tracks = pd.read_csv('tracks.csv')
df_tracks.head()
```

```python
# Checking for null values
pd.isnull(df_tracks).sum()
```

```python
# Getting information on our dataset
df_tracks.info()
```

```python
# Find the top 10 least favorite songs in spotify
sorted_df = df_tracks.sort_values('popularity', ascending=True)
sorted_df.head(10)
```

```python
# Getting descriptive statistics of our dataset
df_tracks.describe().transpose()
```

```python
# Top ten songs with popularity greater than 90
most_popular = df_tracks.query('popularity > 90', inplace = False).sort_values('popularity', ascending=False)
most_popular[:10]
```

```python
# Another method
popular = df_tracks[df_tracks['popularity'] > 90].sort_values('popularity', ascending=False)
popular.head(10)
```

```python
df_tracks.set_index('release_date', inplace=True) # inplace=True means your original dataset wont be changed
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
#filtering the dataset
df_tracks[['artists']].iloc[18]
```

```python
# Converting our duration to seconds instead of milliseconds
df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x: round(x/1000))
df_tracks.drop('duration_ms', inplace=True, axis=1)
df_tracks.duration.head()
```

```python
#Who has the most monthly listeners on spotify
```

```python
# A correlation map
corr_df = df_tracks.drop(['key', 'mode', 'explicit'], axis=1).corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df, annot=True,fmt='.1g', vmin=1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor='Black')
heatmap.set_title('Correlation HeatMap Between Variable')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
# creating a sample df
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
len(sample_df)
```

```python
# Creating a regression plot between loudness and Energy
plt.figure(figsize=(10, 6))
sns.regplot(data=sample_df, x='energy', y='loudness', color='c').set(title = "Loudness vs Energy Correlation")
```

```python
# Creating a regression plot between popularity and acousticness
plt.figure(figsize=(10, 6))
sns.regplot(data=sample_df, x='acousticness', y='popularity', color='b').set(title = "Popularity vs Acousticness Correlation")
```

```python
# create a new column called year release date
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
years.head()
```

```python
#create a distribution plot to visualize the total number of songs since 1922
sns.displot(years, discrete=True, aspect=2, height=5, kind='hist').set(title='Number of songs per year')
```

```python
# a bar plot of Years against Duration of songs
total_dr = df_tracks.duration
fig_dims = (18, 7) # dimensions
fig, ax = plt.subplots(figsize=(fig_dims))
fig = sns.barplot(x=years, y=total_dr, 
                  ax=ax, errwidth=False).set(title='Year vs Duration')
plt.xticks(rotation=90)
plt.show()
```

```python
# a lineplot of Years against Duration of songs
sns.set_style(style='whitegrid')
fig_dims = (10, 5)
fig, ax = plt.subplots(figsize=fig_dims)
fig = sns.lineplot(x=years, y=total_dr, ax=ax).set(title='Year vs Duration')
# plt.xticks(rotation=60)
plt.show()
```
