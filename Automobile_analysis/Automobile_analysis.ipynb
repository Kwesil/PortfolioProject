```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_file = r"C:\Users\Kwesil's\Desktop\Tech\Ml&AI\Data analytics\projects\Automobile.csv"
df = pd.read_csv(data_file)
df.head()
```

```python
df.info()
```

```python
# removing null values
new_df = df.dropna()
new_df.info()
```

```python
# checking the names of the manufacturers
manufacture_names = new_df['name'].value_counts()
manufacture_names
```

```python
# FInding how the weight of the a car affects the speed and acceleration
weight_automobile_data = new_df.loc[:, ['name', 'weight', 'horsepower', 'acceleration', 'mpg']]
weight_automobile_data
```

```python
# Scatter plot to check the relationship  between the weight, horsepower and fuel efficiency

sns.relplot(x='weight', y='horsepower', data=weight_automobile_data, kind="scatter", hue="mpg")
plt.show()
```

```python
sns.relplot(x='weight', y='acceleration', data=weight_automobile_data, kind="scatter", hue="mpg")
plt.show()
```

```python
# checking the relationship between cylinders, horsepower, acceleration and mpg
cylinders_automobile_data = new_df[['name', 'cylinders', 'horsepower', 'acceleration', 'mpg']]
cylinders_automobile_data.head()
```

```python
# scatterplot to visualize the relationship between cylinders, horsepower and mpg

sns.relplot(x='cylinders', y='horsepower', data=cylinders_automobile_data, kind="scatter", hue="mpg")
plt.show()
```

```python
# scatterplot to visualize the relationship between cylinders, horsepower and mpg

sns.relplot(x='cylinders', y='acceleration', data=cylinders_automobile_data, kind="scatter", hue="mpg")
plt.show()
```

```python
# Top 5 fastest cars in the dataset
top_5_fastest_cars = new_df.sort_values('horsepower', ascending=False).head(5)
top_5_fastest_cars
```

```python
top_5_fastest_cars.plot(x='name', y='horsepower', kind='barh', color='orange')
plt.title('Top 5 Fastest Cars')
plt.xlabel('Horsepower')
plt.ylabel('Car Name')
plt.xticks(np.arange(0, 250, 40))
plt.show()
```

```python
# Top 5 slowest cars in the dataset
top_5_slowest_cars = new_df.sort_values('horsepower', ascending=False).tail(5)
top_5_slowest_cars
```

```python
top_5_slowest_cars.plot(x='name', y='horsepower', kind='barh', color='orange')
plt.title('Top 5 Slowest Cars')
plt.xlabel('Horsepower')
plt.ylabel('Car Name')
plt.xticks(np.arange(0, 60, 10))
plt.show()
```
